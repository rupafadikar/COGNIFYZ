```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("Data_set 2 - Copy.csv")
```

```python
# First 5 rows from Dataset
print("First 5 Rows :")
df.head(5)
```

```python
# 1.Analyze Investment Duration Column
investment_duration = df['Duration'].value_counts()
investment_duration
```

```python
# 2.Calculate Average Duration
```

```python
# add numeric value according to duration because duration column is object type

duration_map = {
    "Less than 1 year": 0.5,
    "1-3 years": 2,
    "3-5 years": 4,
    "More than 5 years": 6 
}

df["Duration_Value"] = df["Duration"].map(duration_map)
df[["Duration", "Duration_Value"]]
```

```python
average_duration = df['Duration_Value'].mean()
print("The average investment duration :",average_duration)
```
