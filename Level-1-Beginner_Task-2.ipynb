# Task-2: Gender Distribution
# Objective: Visualize gender distribution in  the dataset

```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("Data_set 2 - Copy.csv")
```

```python
# 1.Extract Gender Information
print("Total Gender Distribution :\n")
print(df['gender'].value_counts())
```

```python
# 2. Visualization of gender distribution

import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
plt.title("Gender Distribution")
plt.xlabel("Gender Category")
plt.ylabel("Value")
plt.bar(df['gender'].value_counts().keys(),
        df['gender'].value_counts(), color=['green','orange'])
plt.show()
```

```python
# 2. Visualization of gender distribution

import matplotlib.pyplot as plt
plt.title("Gender Distribution")
plt.pie(df['gender'].value_counts(), autopct='%1.2f%%',
        labels=df['gender'].value_counts().keys(), colors=['skyblue','r'])
plt.legend()
plt.show()
```
