```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("Data_set 2 - Copy.csv")
```

```python
# First 5 rows from Dataset
print("First 5 Rows :")
df.head(5)
```

```python
# Add another column to define the except object datatype to numeric datatype column
e_return = {
    "20%-30%" : 2,
    "30%-40%" :  3,
    "10%-20%" :  1
}

df["Exp_Return"] = df["Expect"].map(e_return)
df[["Expect","Exp_Return"]]
```

```python
# add numeric value according to duration because duration column is object type

duration_map = {
    "Less than 1 year": 0.5,
    "1-3 years": 2,
    "3-5 years": 4,
    "More than 5 years": 6 
}

df["Duration_Value"] = df["Duration"].map(duration_map)
df[["Duration", "Duration_Value"]]
```

```python
# 1.Select Relevant Columns
relevant_columns = df[['age', 'Duration_Value', 'Exp_Return']]
relevant_columns
```

```python
# Correlation between them
correlation = relevant_columns.corr()
print("Correlation :", correlation)
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("PAIRPLOT OF CORRELATION")
sns.pairplot(relevant_columns)
plt.show()
```

```python
print("SCATTER PLOT")
plt.figure(figsize=(6, 4))
sns.regplot(x=df["age"], y=df["Duration_Value"], scatter_kws={'alpha':0.5})
plt.title("Age vs Investment Duration")
plt.xlabel("Age")
plt.ylabel("Investment Duration (Years)")
plt.show()
```

```python
print("SCATTER PLOT")
plt.figure(figsize=(6, 4))
sns.regplot(x=df["age"], y=df["Exp_Return"], scatter_kws={'alpha':0.5}, line_kws={"color":"red"})
plt.title("Age vs Expected Returns")
plt.xlabel("Age")
plt.ylabel("Expected Returns (%)")
plt.show()
```
